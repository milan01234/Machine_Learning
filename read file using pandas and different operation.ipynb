```python
import matplotlib.pyplot as plt
import csv
  
x = []
y = []
  
with open('Sales.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        x.append(row[0])
        y.append(int(row[7]))
  
plt.bar(x, y, color = 'g', width = 0.72, label = "total_profit")
plt.xlabel('month')
plt.ylabel('totalprofit')
plt.title('Ages of different persons')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
  
x = []
y = []
  
with open('Sales.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        x.append(row[0])
        y.append(int(row[7]))
  
plt.plot(x,y,marker='o',color='b',linestyle=':')
plt.xlabel('month')
plt.ylabel('totalprofit')
plt.title('Ages of different persons')
plt.legend()
plt.show()
```
